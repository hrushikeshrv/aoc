```python
def valid(string):
    c = string.split(':')
    #lower = string[0]
    lower = c[0].split('-')[0]
    upper = c[0].split('-')[1].split()[0]
    letter = c[0].split('-')[1].split()[1]
    password = c[1].strip()
    l_count = 0
    for char in password:
        if char == letter:
            l_count += 1
    if int(lower) <= l_count <= int(upper):
        return 1
    else:
        return 0
    
def solve(input_string):
    correct = 0
    for password in input_string:
        correct += valid(password)
    
    return correct
```

```python
def valid_2(string):
    c = string.split(':')
    lower = int(c[0].split('-')[0]) - 1
    upper = int(c[0].split('-')[1].split()[0]) - 1
    letter = c[0].split('-')[1].split()[1]
    password = c[1].strip()
    
    if password[lower] == letter and password[upper] != letter:
        return 1
    elif password[lower] != letter and password[upper] == letter:
        return 1
    else:
        return 0

def solve_2(input_string):
    correct = 0
    for password in input_string:
        correct += valid_2(password)
    return correct
```

```python
def check_valid(d):
    valid = True
    if not 1920 <= int(d.get('byr', 0)) <= 2002:
        valid = False
    if not 2010 <= int(d.get('iyr', 0)) <= 2020:
        valid = False
    if not 2020 <= int(d.get('eyr', 0)) <= 2030:
        valid = False
    if d.get('hgt', '0in').endswith('in'):
        if not 59 <= int(d.get('hgt', '0in')[:-2]) <=76:
            valid = False
    if d.get('hgt', '0cm').endswith('cm'):
        if not 150 <= int(d.get('hgt', '0cm')[:-2]) <=193:
            valid = False
    if d.get('ecl', 'none') not in 'amb blu brn gry grn hzl oth'.split():
        valid = False
    if not (len(d.get('pid', 'none')) == 9 and d.get('pid', 'none').isnumeric()):
        valid = False
    color = d.get('hcl', None)
    if color is None or color[0] != '#' or len(color) != 7:
        valid = False
    else:
        for c in color[1:]:
            if c not in '0123456789abcdef':
                valid = False
                break
    
    return valid
    

def solve_2(inp):
    checks = (
        'byr',
        'iyr',
        'eyr',
        'hgt',
        'hcl',
        'ecl',
        'pid',
    )
    valids = 0
    for passport in inp:
        keys = [x.split(':')[0] for x in passport]
        values = [x.split(':')[1] for x in passport]
        
        p_dict = {keys[i]:values[i] for i in range(len(keys))}
        if check_valid(p_dict):
            valids += 1
    
    return valids
```
